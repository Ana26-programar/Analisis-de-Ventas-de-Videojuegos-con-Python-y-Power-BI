#Valores nulos

Contar nulos por columna

Filtrar filas con editorial nula

Eliminar filas con editorial nula

Verificar que no quedan nulos

```python
#Contar nulos por columna
df.isnull().sum()
#Filtrar filas con editorial nula
df[df["Editorial"].isnull()]
#Eliminar filas con editorial nula
df = df.dropna(subset=["Editorial"])
#Verificar que no quedan nulos
df = df.dropna()
df.isnull().sum()
```

#Columnas nuevas

Crear columna Ventas_Regional_Total (NA+EU+JP+Otros)

Crear columna Diferencia_Global_Regional

Crear columna Decada a partir de Año

Crear columna booleana Exito (Ventas Global > 1)

```python
#Crear columna Ventas_Regional_Total (NA+EU+JP+Otros)
df["Ventas_Regional_Total"] = df["Ventas NA"] + df['Ventas EU'] + df['Ventas JP'] + df['Ventas Otros']
#Crear columna Diferencia_Global_Regional
df["Diferencia_Global_Regional"] = df['Ventas Global'] - df["Ventas_Regional_Total"]
df = df.drop(columns=["Diferencia_Global_Regional"])
#Crear columna Decada a partir de Año
df["Decada"] = (df["Año"] // 10) * 10
#Crear columna booleana Exito (Ventas Global > 1)
df["Exito"] = df["Ventas Global"] > 1
df
```

#GroupBy (CLAVE)

Ventas globales por plataforma

Ventas globales por género

Ventas globales por editorial

Ventas globales por año

Ventas globales por década

Ventas globales por género y editorial

```python
#Ventas globales por plataforma
df.groupby("Plataforma")["Ventas Global"].sum()
#Ventas globales por género
df.groupby("Genero")["Ventas Global"].sum()
#Ventas globales por editorial
df.groupby("Editorial")["Ventas Global"].sum()
#Ventas globales por año
df.groupby("Año")["Ventas Global"].sum()
#Ventas globales por década
df.groupby("Decada")["Ventas Global"].sum()
#Ventas globales por género y editorial
df.groupby(["Genero", "Editorial"])["Ventas Global"].sum()
```

#Ranking

Top 5 plataformas por ventas globales

Top 5 géneros en Europa

Top 5 editoriales en Norteamérica

Ranking de juegos por ventas globales

```python
#Top 5 plataformas por ventas globales
df.groupby("Plataforma")["Ventas Global"].sum().sort_values(ascending = False).head(5)
#Top 5 géneros en Europa
df.groupby("Genero")["Ventas EU"].sum().sort_values(ascending = False).head(5)
#Top 5 editoriales en Norteamérica
df.groupby("Editorial")["Ventas NA"].sum().sort_values(ascending = False).head(5)
#Ranking de juegos por ventas globales
df.groupby("Nombre")["Ventas Global"].sum().sort_values(ascending = False).head(5)
```

#Exportación

Exportar dataframe limpio a CSV

Exportar tabla agregada (ventas por plataforma)

Exportar top 10 juegos a CSV

```python
#Exportar dataframe limpio a CSV
df.to_csv("ventas_videojuegos_limpio.csv", index=False)

#Exportar tabla agregada (ventas por plataforma)
ventas_plataforma = (
    df.groupby("Plataforma")["Ventas Global"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)
ventas_plataforma.to_csv("ventas_por_plataforma.csv", index=False)

#Exportar top 10 juegos a CSV
top_10_juegos = (
    df.sort_values(by="Ventas Global", ascending=False)
      .head(10)
)
top_10_juegos.to_csv("top_10_juegos.csv", index=False)
```
